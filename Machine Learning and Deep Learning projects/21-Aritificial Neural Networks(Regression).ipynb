#IMPORTING THE LIBRARIES

```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

#IMPORTING THE DATASET

```python
dataset = pd.read_excel('Folds5x2_pp.xlsx')
X = dataset.iloc[:, :-1].values #columns 0,1,2 index does not provide any impact for the model. so we start from 3rd column and exclude the last column
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

#Splitting dataset into train and test cell

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

#Part 2-Build the ANN

Initializing the ANN

```python
ann=tf.keras.models.Sequential()
```

Adding the input layer and the first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu')) #6 neurons in the hidden layer
```

Adding the second hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

Adding the ouput layer

```python
ann.add(tf.keras.layers.Dense(units=1)) #no activation function for regression
```

#Training the ANN

Compiling the ANN

```python
ann.compile(optimizer='adam',loss='mean_squared_error',metrics=['accuracy'])
#optimizer updates the weight through stochastic gradient descent
#loss to compute difference between predictions and real
#metrics for accuracy
#loss=mean squared error for regression
```

Training the ANN on the training set

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
#batch size refers to number of predictions vs  the real results
#epochs refer to number of times the neural network is trained
```

#Making the predictions and evaluating the model

#Predicting the Test set results

```python
y_pred = ann.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

#Making the confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
