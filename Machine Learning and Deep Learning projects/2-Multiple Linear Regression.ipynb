#IMPORTING THE LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

#IMPORTING THE DATASET

```python
dataset=pd.read_csv('50_Startups.csv') #this makes a dataframe
#features are the columns and dependent variable is the last column of the dataset

X=dataset.iloc[: , :-1].values 
#iloc stands for locate indexes and takes indexes of the columns and rows. here we will take  [rows,columns]  . 
#-1 index to exclude the last column

y=dataset.iloc[: , -1].values  
#-1 index is mentioned to only take the last column 
```

```python
print(X)
```

#Encoding the categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

#creating an object of the ColumnTransformer class
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])],remainder='passthrough')
X=np.array(ct.fit_transform(X))
```

```python
print(X)
```

#SPLIT THE DATASET

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

### **IN MULTIPLE LINEAR REGRESSION NO NEED TO APPLY FEATURE SCALLING imp**

#Training the multiple linear regression model on the training set

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)

#no need to worry about the dummy variable becoz this class takes cares according to the p values and selects the best
```

#Predicting the test set

```python
y_pred=regressor.predict(X_test) 
np.set_printoptions(precision=2) #will display any numerical values only two decimals after comma
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
#concatenation of two vertical vectors (predict vs actual from test set)
```

```python
plt.scatter(range(len(y_pred)), y_pred, color='blue', label='Predicted')
plt.scatter(range(len(y_test)), y_test, color='red', label='Actual')
plt.xlabel('Data Points')
plt.ylabel('Dependent Variable')
plt.title('Predicted vs Actual')
plt.legend()
plt.show()
```

#Evaluate the performance using R2 squared

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
