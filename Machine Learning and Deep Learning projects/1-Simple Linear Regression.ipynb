IMPORTING THE LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

IMPORTING THE DATASET

```python
dataset=pd.read_csv('Salary_Data.csv') #this makes a dataframe
#features are the columns and dependent variable is the last column of the dataset

X=dataset.iloc[: , :-1].values 
#iloc stands for locate indexes and takes indexes of the columns and rows. here we will take  [rows,columns]  . 
#-1 index to exclude the last column

y=dataset.iloc[: , -1].values  
#-1 index is mentioned to only take the last column 
```

Split the dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=1/3,random_state=1)
```

Training the simple linear regression model on the training set

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

Predicting the Test set Results

```python
y_pred=regressor.predict(X_test) 
#y_pred contains the predicted salaries
#y_test contains the actual salaries

#so that we will predict the salary from the years of exp
#that is why only x test is given to implement the years
```

Visualizing the training set results

```python
plt.scatter(X_train,y_train,color='red') #this plots the actual salaries 
plt.plot(X_train,regressor.predict(X_train),color='blue') #gives predicted salaries of the x_train
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

Visualizing the test set results

```python
plt.scatter(X_test,y_test,color='red') #this plots the actual salaries 
plt.plot(X_train,regressor.predict(X_train),color='blue') #gives predicted salaries of the x_test
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()

#predicted salaries of the test set will be on the same regression line as that of the training set
#because linear regression has a unique equation.Hence we do not need to change the plt.plot
```

Making a single prediction(for example the salary of an employee with 12 years of experience)

```python
print(regressor.predict([12]))
```

Value of the feature was put in a double pair of square brackets because 'predict' method always expects a 2d array

#Final linear regression equation with coefficient values

```python
print(regressor.coef_)
print(regressor.intercept_)
```
