
 IMPORTING THE LIBRARIES

```python
#numpy is used to work with arrays
#matplotlib is used to work to plot charts
#pandas to import the dataset and also create matrix of features
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

IMPORTING THE DATASET


```python
dataset=pd.read_csv('Data.csv') #this makes a dataframe
#features are the columns and dependent variable is the last column of the dataset

X=dataset.iloc[: , :-1].values 
#iloc stands for locate indexes and takes indexes of the columns and rows. here we will take  [rows,columns]  . 
#-1 index to exclude the last column

y=dataset.iloc[: , -1].values  
#-1 index is mentioned to only take the last column 
```

```python
print(X)
```

```python
print(y)
```

TAKING CARE OF MISSING DATA

```python
from sklearn.impute import SimpleImputer 
imputer=SimpleImputer(missing_values=np.nan,strategy='mean') #object of the simple imputer class
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])

#sklearn.impute is a module within the scikit-learn library, which provides a set of classes for handling missing values in datasets.
#update the missing salary with the average salary 
```

```python
print(X)
```

ENCODING CATEGORICAL DATA

ENCODING THE INDEPENDENT VARIABLE

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

#creating an object of the ColumnTransformer class
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
X=np.array(ct.fit_transform(X))
#to apply one hot encoding
```

```python
print(X)
```

  ENCODING THE DEPENDENT VARIABLE

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder() #label refers to the dependent variable. it refers to the output
y=le.fit_transform(y)
```

```python
print(y)
```

SPLITTING THE DATASET INTO TRAINING AND TEST SET

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
print(X_train) #8 observations or 8 customers randomly
```

```python
print(X_test) #2 obs. or 2 customers from 0.2 test size
```

```python
print(y_train)
```

```python
print(y_test)
```

#feature scalling is done after the train and test split to prevent information leakage

FEATURE SCALLING

```python
#feature scalling is done to avoid one feature to not dominate the other feature
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train[:,3:]=sc.fit_transform(X_train[:,3:])
X_test[:,3:]=sc.transform(X_test[:,3:])
```
